Donde $L_0$ indica la línea vacía y las $\overline{L_i}$ indican los complementos de cada línea.

Como podemos observar de la tabla, $d_{min} = 3$, por lo que sabemos del teorema de capacidad de detección y corrección de errores que podemos corregir como máximo $1$ error usando este código $(d_{min}>2t+s)$, con $t=1$ y $s=0$.

Así entonces veamos una forma de hacer la decodificación de una palabra $w$, dependiendo del número de unos, $u(w)$, en dicha palabra:
* Si $u(w) = 0$, decodificamos el mensaje $0000$ (correspondiente a la línea vacía).
* Si $u(w) = 1$, debido a que $d_{min} = 3$, solo podemos tener que el posible mensaje es $0000$ (correspondiente a la línea vacía).
* Si $u(w) = 2$, sabemos que cada 1 en $w$ tiene asociado un punto, así que hay dos puntos asociados a $w$. De la teoría de la geometría proyectiva sabemos que dos puntos siempre son colineales y que dicha línea es única, así que decodificamos con el mensaje asociado a dicha línea.
* Si $u(w) = 3 $, así que tenemos $3$ puntos asociados a $w$. Hay dos posibles opciones. La primera de ellas es que no haya ningun error, lo que se interpretará geométricamente como que los 3 puntos son colineales y por tanto decodificamos con el mensaje asociado a dicha palabra. La segunda opción es que haya 1 error. Geométricamente esto significa que los puntos forman un triángulo. Dicho triángulo puede convertirse en un cuadrilátero al agregar otro punto (es decir, al cambiar un $0$ por un $1$). Este punto se obtiene al descartar los puntos que son colineales con respecto a 2 de los 3 puntos iniciales. Tenemos entonces que hay ${3 \choose 2}=3$, así que podemos eliminar $3$ puntos de los $4$, dejándonos un único punto, al que cambiaremos su valor en la palabra $w$ por un 1. Ahora, notemos que una palabra con $4$ unos está asociada con una línea complementaria, así que decodificamos con el mensaje asociado a dicha línea. Otra interpretación de es que debemos encontrar la línea que no contiene a ninguno de los $3$ puntos asociados con la palabra $w$ y hacer su complemento; la línea obtenida será la usada para la decodificación de $w$.
* Si $u(w)=4,5,6$ o $7$, simplemente hacemos el complemento de $w$, la cual tendrá $u(\overline{w})= 3,2,1$ o $0$ (respectivamente), que es un caso anterior. Decodificamos $\overline(w)$ y hacemos el complemento de dicha decodificación.

Veamos algunos ejemplos para ilustrar
* Si recibimos $w=0010001$. Esto significa que tenemos asociados los puntos $P_3$ y $P_7$. Mirando en los gráficos, tenemos que ambos puntos están sobre la línea $L_2$, y el punto faltante es $P_1$, por lo que modificamos $w$ obteniendo $1010001$ y mirando en la tabla para decodificar (o simplemente mirando los últimos 4 dígitos), obtenemos que el mensaje decodificado es $0001$.
* Si recibimos $w=1001010$, tenemos asociados los puntos $P_1$, $P_4$ y $P_6$. Mirando el plano de Fano, vemos que estos puntos son colineales y su línea asociada es $L_4$, lo que significa que el código no tiene ningun error y el mensaje decodificado es $1010$.
* Si recibimos el mensaje $w=1110000$, caemos en el caso $u(w)=3$. Tenemos asociados los puntos $P_1$, $P_2$ y $P_3$. Si vemos en el plano de Fano, estos puntos no son colineales, por lo cual sabemos que hay un error. Podemos buscar la línea que no contiene a ninguno de estos puntos, que en este caso es $L_6$, por lo cual sabemos que la línea asociada a $w$ es $\overline{L_6}$. La decodificación entonces nos da el mensaje $0010$. Esto es difícil de ver para el computador, así que usemos el otro método. Sabemos que si sumamos dos puntos, obtenemos otro punto colineal a ellos. Así que: $P_1+P_2=001+010=011 = P_5$, por lo cual descartamos a $P_5$. También tenemos que $P_1 + P_3 =001+100=101=P_7$, descartando a $P_7$. Por último $P_2+P_3 = 010++100 = 110 = P_4$. Al final tenemos que tenemos que cambiar el 0 de la posición 6 de $w$, obteniendo la palabra $1110010$ que corresponde con la línea complementaria a $L_6$ y obtenemos el mensaje decodificado $0010$.

Vamos a mostrar a continuación un código en Python para decodificar una palabra recibida y corregirla si es el caso.

```
import numpy

def generarMensajeAleatorio(n):
  word=numpy.random.choice([0,1],size=7*n)
  texto=""
  for l in word:
    texto += str(l)
  return texto

def decodificarMensaje(mensaje):
  n=len(mensaje)//7
  decodificacion=""
  for i in range(0,n):
    palabra=mensaje[i*7:(i+1)*7]
    u=palabra.count("1")
    if u<2:
      decodificacion+="0000"
    print(u)
mensajeRecibido=generarMensajeAleatorio(20)
print(mensajeRecibido)
decodificarMensaje(mensajeRecibido)

```
